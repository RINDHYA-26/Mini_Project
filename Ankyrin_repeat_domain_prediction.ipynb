```python
def convert_to_fasta(input_file, output_file):
    with open(input_file,'r') as infile, open(output_file,'w') as outfile:
        for i, line in enumerate(infile, 1):
            sequence = line.strip()
            outfile.write(f">{i}\n{sequence}\n")

# Replace 'input_sequences.txt' and 'output_sequences.fasta' with your actual file names
convert_to_fasta('pos-SEQ.txt','positive-sequences.fasta')
```

```python
pos_arr=[[0 for i in range(0,21)] for j in range(0,102)]
AA=['A','C','D','E','F','G','H','I','K','L','M','N','P','Q','R','S','T','V','W','Y']
count=0
with open('pos_cdhit.txt','r') as infile:
        for i, line in enumerate(infile, 1):
          if(i%2==0):
            sequence = line.strip()
            for j in range(0,20):
              if(AA[j] in sequence):
                pos_arr[count][j]=sequence.count(AA[j])
            pos_arr[count][20]=len(sequence)
            count+=1
```

```python
neg_arr=[[0 for i in range(0,21)] for j in range(0,121)]
AA=['A','C','D','E','F','G','H','I','K','L','M','N','P','Q','R','S','T','V','W','Y']
count=0
with open('neg_cdhit.txt','r') as infile:
        for i, line in enumerate(infile, 1):
          if(i%2==0):
            sequence = line.strip()
            for j in range(0,20):
              if(AA[j] in sequence):
                neg_arr[count][j]=sequence.count(AA[j])
            neg_arr[count][20]=len(sequence)
            count+=1
```

```python
properties=pd.DataFrame()
for k in range(0,49):
  c0=[]
  for i in range(len(data)):
    K0=[]
    for j in range(0,20):
      if (data[i][j]!=0):
        K0.append(data[i][j]*norm_prop[AA[j]].iloc[k])
    c0.append(round(sum(K0)/data[i][20],2))
  c0=pd.DataFrame(c0)
  properties[norm_prop['Property'].iloc[k]]=pd.DataFrame(c0)
properties['class']=pd.DataFrame(feature_class)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Assuming you have a pandas DataFrame named X with 49 columns and 1800 rows

# Perform PCA on the dataset
pca = PCA()
pca.fit(X)

# Get the explained variances and cumulative explained variances
explained_variances = pca.explained_variance_
explained_var_ratio = pca.explained_variance_ratio_
cumulative_explained_var = np.cumsum(explained_var_ratio)

# Plot the scree plot
plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, len(explained_variances) + 1), cumulative_explained_var, marker='o')
plt.title('Scree Plot')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Perform k-fold cross-validation
n_splits = 5  # Number of folds for cross-validation
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Initialize lists to store true labels and predicted probabilities
true_labels = []
predicted_probabilities = []
predicted_labels = []

# Perform k-fold cross-validation and collect true labels and predicted probabilities
for train_index, test_index in kf.split(X_train):
    X_train1, X_test1 = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train1, y_test1 = y_train.iloc[train_index], y_train.iloc[test_index]

    # Fit the random forest model on the training data
    rf.fit(X_train1, y_train1)

    # Predict probabilities on the test data
    y_prob = rf.predict_proba(X_test1)[:, 1]

    # Collect true labels and predicted probabilities
    true_labels.extend(y_test1)
    predicted_probabilities.extend(y_prob)

# Calculate the false positive rate (FPR), true positive rate (TPR), and threshold values for the ROC curve
fpr, tpr, thresholds = roc_curve(true_labels, predicted_probabilities)

# Calculate the area under the ROC curve (AUC)
roc_auc = auc(fpr, tpr)
# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, color='b', label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='r', linestyle='--', label='Random')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()

# Calculate the predicted labels based on the predicted probabilities
predicted_labels = [1 if prob >= 0.5 else 0 for prob in predicted_probabilities]

# Calculate the accuracy
accuracy = (sum(np.array(true_labels) == np.array(predicted_labels)) / len(true_labels))*100
round(accuracy,2)
```
