```python
def convert_to_one_liner(input_filename, output_filename):
    with open(input_filename, 'r') as infile:
        sequences = infile.read().split('>')

    # Remove empty first element from the split
    sequences.pop(0)

    with open(output_filename, 'w') as outfile:
        for sequence in sequences:
            lines = sequence.split('\n', 1)
            header = lines[0]
            sequence_data = ''.join(line.strip() for line in lines[1].split('\n'))
            outfile.write(f'>{header}\n{sequence_data}\n')

# Usage
convert_to_one_liner('final_pos_seq.fasta', 'positive.fasta')
convert_to_one_liner('final_neg_seq.fasta', 'negative.fasta')
```
